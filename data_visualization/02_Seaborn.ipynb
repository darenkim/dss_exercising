```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0,14, 100)

y1 = np.sin(x)
y2= 2*np.sin(x+0.5)
y3= 3*np.sin(x+1.0)
y4= 4*np.sin(x+1.5)
```

```python

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x, y3, x,y4);
```

```python
sns.set_style("white")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x, y3, x,y4)
sns.despine();
```

```python
sns.set_style("dark")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x, y3, x,y4)
sns.despine(offset= 0);
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x, y3, x,y4);
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x, y3, x,y4)
sns.despine(offset=10); #맨 첫 칸을 띄어줌
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x=tips['total_bill']);
```

```python
sns.boxplot(x='day', y = 'total_bill', data = tips);
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='day', y = 'total_bill',hue = 'smoker', data = tips, palette='Set3');
```

```python
sns.swarmplot(x = 'day', y='total_bill', data=tips)
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x = 'day', y='total_bill', data=tips)

sns.boxplot(x='day', y = 'total_bill', data = tips, palette='Set3');
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, height=7);
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips,hue = 'smoker', height=7);
```

```python
flights =sns.load_dataset('flights')
flights.head()
```

```python
import pandas as pd
flights = pd.pivot_table(flights, 'passengers', 'month', 'year',aggfunc=np.sum)
flights.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt= 'd');
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt= 'd', cmap= 'YlGnBu');
```
