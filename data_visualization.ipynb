```python
crime = pd.read_csv('./data/2017_crime.csv', encoding='euc-kr')
crime.head()
```

```python
crime.info()
```

```python
crime_st = crime.pivot_table(crime, index=['구분'], columns=['죄종', '발생검거'], aggfunc = [np.sum])

crime_st.head()
```

```python
crime_st.columns
```

```python
crime_st.columns = crime_st.columns.droplevel([0,1])
crime_st.columns
```

```python
crime_st.head()
```

```python
tmp = crime_st.columns.get_level_values(0)+ crime_st.columns.get_level_values(1)
tmp
```

```python
crime_st.columns = tmp
crime_st.head()
```

```python
station_names = ['혜화', '남대문', '중부', '종암', '서부', '방배', '수서']
gu_names = ['종로구', '중구', '중구', '성북구', '은평구', '서초구', '강남구']
```

```python
tmp_gu = []

for idx, _ in crime_st.iterrows():
    if idx in station_names:
        tmp_gu.append(gu_names[station_names.index(idx)])
    else:
        tmp_gu.append(idx+'구')
```

```python
station_names.index('중부')
```

```python
crime_st['구'] = tmp_gu
crime_st.head()
```

```python
crime_gu = pd.pivot_table(crime_st, index='구', aggfunc= np.sum)
crime_gu.head()
```

```python
num=['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']

den=['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']
target=['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_gu[target] = crime_gu[num].div(crime_gu[den].values)*100
```

```python
crime_gu.head()
```

```python
crime_gu = crime_gu.drop(columns = num)
crime_gu.head()
```

```python
crime[crime_gu[target] > 100] == 100
crime_gu.head()
```

```python
crime_gu.rename(columns = {'강간발생' : '강간', '강도발생' : '강도', '살인발생': '살인', '절도발생' :'절도', '폭력발생':'폭력'}\
               ,inplace = True)

crime_gu.head()
```

```python
col = ['살인', '강도', '강간', '절도', '폭력']
crime_gu_norm= crime_gu[col] / crime_gu[col].max()
crime_gu_norm.head()
```

```python
col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_gu_norm[col2]=crime_gu[col2]
crime_gu_norm.head()
```

```python
col = ['강간', '강도', '살인', '절도', '폭력']
crime_gu_norm['범죄'] = np.mean(crime_gu_norm[col], axis =1)
crime_gu_norm.head()
```
