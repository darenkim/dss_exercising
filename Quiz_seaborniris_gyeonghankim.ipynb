```python
# enviroment setting 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# load dataset
iris_df = sns.load_dataset('iris')
```

# Quiz 1. make pivot table 
- iris_df 데이터 프레임을 활용하여서 종에 대한 피봇 테이블을 만들어 주세요!
- 아래와 같은 결과가 나오면 됩니다
- 값들은 평균값들이 나왔으면 좋겠습니다!

```python
iris_df.groupby('species')['petal_length', 'petal_width', 'sepal_length', 'sepal_width'].mean()

#결과는 같습니다....
```

```python
iris_df.pivot_table(index = ['species'],aggfunc = np.mean)
```

```python
iris_df.pivot_table(index=['species'], aggfunc=np.sum)
```

```python
iris_df.pivot_table(iris_df,'species', aggfunc=np.mean)
```

# Quiz 2. make boxplot
- iris_df를 활용해서 boxplot을 그려주세요!
- x_values = sepal_length , y_values  = petal_width 
- 종으로 분류 해주세요!   

```python
plt.figure(figsize = (14,8))
sns.boxplot(x='sepal_length', y='petal_width', data= iris_df, hue = 'species', palette='Set3');
```

#  Quiz 3. make lmplot
- iris_df 를 활용해서 lmplot을 그래주세요!
- x_values = sepal_length , y_values  = petal_width
- lmplot은 종에 따라 구별되게 해주세요!   

```python
sns.set_style("darkgrid")
plt.figure(figsize = (12,8))
sns.lmplot(x='sepal_length', y='petal_width', data=iris_df,hue = 'species');
```

#  Quiz 4. make pairplot
- iris_df를 활용하여서 pairplot을 그려주세요!
- x_values = sepal_length','sepal_width, y_values  = petal_width
- 종으로 분류 해주세요!    

```python
plt.figure(figsize = (14,8))

sns.pairplot(iris_df, x_vars=['sepal_length', 'sepal_width'],
            y_vars=['petal_length'], hue='species', height = 4);
```
